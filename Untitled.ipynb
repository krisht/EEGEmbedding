```python
start, stop = raw.time_as_index([100, 115])
data, times = raw[:36, start:stop]
print(data.shape)
print(times.shape)
print(times.min(), times.max())
```

```python
picks = mne.pick_types(raw.info, meg=True, eeg=True, stim=False, ecg=False, emg=False, exclude = 'bads');
print(picks)

data, times = raw[picks[:10], start:stop]
```

```python
import matplotlib.pyplot as plt
import plotly.plotly as py
plt.plot(times, data.T)
plt.xlabel('time(s)')
plt.ylabel('EEG data (T)')

update = dict(layout=dict(showlegend=True), data=[dict(name=raw.info['ch_names'][p]) for p in picks[:10]])
py.iplot_mpl(plt.gcf(), update=update)
```
